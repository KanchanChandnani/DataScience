#Loading The Data

```python
import pandas as pd
import numpy as np

#Reading DataSet from CSV
dataframe = pd.read_csv('http://www.cs.stonybrook.edu/~skiena/591/hw1/country-data.csv', delimiter=',')
```

#Handling Missing Values

```python
#Replacing 'unknown' with NaN and 1.78% in Military Expenditure with 1.78
dataframe = dataframe.replace('unknown', np.nan)
dataframe['Military Expenditure/GDP'] = dataframe['Military Expenditure/GDP'].replace('1.78%','1.78')
dataframe.iloc[:,7:21] = dataframe.iloc[:,7:21].astype(float)
dataframe = dataframe.fillna(dataframe.mean(axis = 0))
```

#Question 1

Data Map

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.io.shapereader as shpreader
import numpy as np

fig,ax = plt.subplots(1)

shapename = 'admin_0_countries'
countries_shp = shpreader.natural_earth(resolution='110m',
                                        category='cultural', name=shapename)

colors = ["#F1EEF6", "#D4B9DA", "#C994C7", "#DF65B0", "#DD1C77", "#980043"]

dataframe['Literacy Rate (%)'] = dataframe['Literacy Rate (%)'].astype(float)
Map = dataframe['Country Name']
ax = plt.axes(projection=ccrs.PlateCarree())
Map = Map.tolist()
Map2 = dataframe['Literacy Rate (%)'].tolist()

for country in shpreader.Reader(countries_shp).records():
    if country.attributes['name_long'] in Map:
        idx = Map.index(country.attributes['name_long'])
        literacy = Map2[idx]
        idx = min(int(literacy/30), 5)
        ax.add_geometries(country.geometry, ccrs.PlateCarree(),
                      facecolor=colors[idx],
                      label=country.attributes['name_long'])

plt.title('Literacy Rate Data Map')
fig.savefig("DataMap.png")
plt.show()
```

 <img src="https://raw.githubusercontent.com/KanchanChandnani/DataScience/master/DataMap.png">

The plot uses various shades of pink and purple in order to depict the literacy rate of the countries of the world. The brighter the shade (pink) of the country the higher the literacy rate is. The data map provides an interesting correlation between the development of the country and it's literacy rate. As seen in the map, developed countries seem to have a higher literacy rate as compared to developing countries. A greater part of the world has a high literacy rate. Most of the countries with a lower literacy rate lie in Africa and Asia. Also, countries with extremely low literacy rate are land locked, as depicted by the map.

```python
import prettyplotlib as ppl
import matplotlib as mpl
import matplotlib.pyplot as plt
import brewer2mpl

fig, ax = plt.subplots(1)
set2 = brewer2mpl.get_map('Set2', 'qualitative', 8).mpl_colors
x_range = [0,50,60,70,80,90]
ax.set_ylim([0,900000])
ax.set_xlim([45,90])
for i in range(5):
    color = set2[i]
    temp = dataframe[(dataframe['Life Expectancy']>x_range[i]) & (dataframe['Life Expectancy']<=x_range[i+1])]
    ppl.scatter(ax,temp['Life Expectancy'],temp['Health Expenditure/GDP']*temp['GDP (PPP) in US $']/10000000,label="LE<"+str(x_range[i+1]),alpha=0.8,color=color)

ppl.legend(ax)
plt.xlabel('Life Expectancy (LE)')
plt.ylabel('Health Expenditure in 10 Million US $')
_ = ax.set_title('Life Expectancy vs Health Expenditure/GDP')
fig.savefig('scatter.png')
plt.show()
```

The scatter plot shows how Life Expectancy of the countries varies with Health Expenditure. The plot does not show a direct correspondence between Health Expenditure and Life Expectancy. On an average, countries with low life expectancy (LE < 55) spend less on health. As for countries with life expectancy in the range 55 - 70, it is not necesarily true that they spend proportionally on health.
```python
import sklearn
from sklearn import linear_model

X = pd.concat([dataframe['Internet Users (circa 2009)'], dataframe['Labor Force'],dataframe['Health Expenditure/GDP'], dataframe['Military Expenditure/GDP']],axis = 1)
Y = dataframe['GDP (PPP) in US $']/dataframe['Population']
features_train, features_test, target_train, target_test = sklearn.cross_validation.train_test_split(X,Y, test_size=0.33)

linreg = linear_model.LinearRegression()
linreg.fit(features_train, target_train)
Output = pd.Series(linreg.predict(X))
difference = Output-Y
Print = pd.concat([dataframe['Country Name'],Y,Output,difference],keys=['Country Name','Actual Value','Predicted Value','Difference'],axis = 1)
#Countries Most Above Forecast
Print.sort(columns = 'Difference').head()
```

```python
import scipy as sp

Ranking_Function = dataframe['GDP (PPP) in US $']/dataframe['Population'] + dataframe['Life Expectancy'] + dataframe['Literacy Rate (%)']
Output = pd.Series(len(dataframe)+1-sp.stats.rankdata(Ranking_Function))
Print = pd.concat([dataframe['Country Name'],Output,dataframe['GDP (PPP) in US $']/dataframe['Population'],dataframe['Life Expectancy'], dataframe['Literacy Rate (%)']],axis=1,keys=['Country Name','Rank','GDP Per Capita','Life Expectancy','Literacy Rate'])
Print = Print.sort(columns = 'Rank')
Print
```

I have used GDP per Capita, Literacy Rate and Life Expectancy as a measure of Social welfare. Social Welfare is highly dependent on the income of the individuals of a country. GDP has an income and expenditure view, and thus can be used as a measure of income. Apart from that, social welfare is a measure of the utilities (economic welfare) provided for the people of the country are which can be denoted as a function of Life Expectancy and Literacy rate. Health and Education are the most important utilities provided to the country's citizens. Thus, I used these measures for the social welfare function.

The social welfare function gives Monaco, Qatar, Liechtensten, Luxemborg and Singapore the top 5 ranks and Central African Republic, Zimbabwe, Burundi, Somalia and Congo the lowest 5 ranks. This is a plausible outcome of a social welfare function. 

Other factors such as Health Expenditure could be used to make the function more accurate. Intuitively, the more the country spends on health, the better economic and social welfare it is providing its citizens. Also,weights can be used for each measure in order to give more importance to a certain aspect of social welfare. For example, the Bergson–Samuelson social welfare function considers the individual income as the most important measure of social welfare.

In conclusion, I feel that my social welfare function measures the important characteristics of a country that would provide desirable social welfare to it's citizens. It can be used to give a more accurate rank by providing weights to the various measures of the social welfare function.

```python
import pandas as pd
import scipy as sp

dataframe['Population Density'] = dataframe['Population']/dataframe['Area (sq km)']
dataframe['GDP Per Capita'] = dataframe['GDP (PPP) in US $']/dataframe['Population']

NewDataFrame = pd.concat([dataframe['Population Density'],dataframe['GDP Per Capita'],
                          dataframe['Literacy Rate (%)'],dataframe['Area (sq km)']],axis=1)

distance = pd.DataFrame(sp.spatial.distance.squareform(sp.spatial.distance.pdist(NewDataFrame, metric='cityblock')))
distance = distance.replace(0,np.nan)
Minimum_Maximum= pd.DataFrame(index=range(0,197), columns=['Country Name','Minimum Distance','Maximum Distance'])
Minimum_Maximum
for i in range(0,197):
    countryIndexMin=distance.iloc[i,:].idxmin()
    countryIndexMax=distance.iloc[i,:].idxmax()
    Minimum_Maximum.iloc[i,0] = dataframe.iloc[i,0]
    Minimum_Maximum.iloc[i,1] = dataframe.iloc[countryIndexMin,0]
    Minimum_Maximum.iloc[i,2] = dataframe.iloc[countryIndexMax,0]
Minimum_Maximum
```

Their physical and economic characteristics are similar in nature. The measure also shows that Turkey and Pakistan are similar in nature, which they are in terms of physical characteristics. Turkey has a higher GDP per capita, which makes it economically better than Pakistan. Liechtenstein and Luxembourg are shown to be most similar even though there is a vast distance between their GDPs, as their GDP per capita is similar.

Thus, the distance measure gets Economic and Physical similarities right.
